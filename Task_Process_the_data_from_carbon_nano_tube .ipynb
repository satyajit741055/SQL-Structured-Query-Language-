# Checking Database 
1. Taking name of database from user 
2. checking the Database exists or not 
3. if it is exists then printing "Database Exists"
4. if not exists then create Database 

```python
mydb = connection.connect(host='localhost',user = 'root',passwd='root',use_pure = True)
database_name = input('enter Database name: ')
cursor = mydb.cursor()
cursor.execute('show databases')
data = cursor.fetchall()
database_list = []
for i in data:
    j = list(i)
    database_list.append(j[0])
try :
    if database_name in database_list:
        print(f'{database_name} Database Exists')
    else:
        print('Creating Database')
        cursor = mydb.cursor()
        cursor.execute(f'create database {database_name}')
        print(f'{database_name} created')
except Exception as e:
    print("database Exist No need to create ")
```

# Preporocessing Data 
1. We have to preporcess the data from above csv file 
2. Mainly i have replaced ";" to "," so string can parse the values to database 


```python
data = open('carbon_nanotubes.csv','r')
f = data.readlines()
data.close()
data2 = open('carbon_nanotubes.csv','w')
for line in f:
    line = line.replace(';',',')
    data2.write(line)
data2.close()
```

# Inserting Preprocessed Values to Database from CSV file 

```python
import csv
mydb = connection.connect(host='localhost',database = 'carbon_nanotubes',user = 'root',passwd='root',use_pure = True)
cursor = mydb.cursor()
with open('carbon_nanotubes.csv',"r") as data :
    data_csv = csv.reader(data,delimiter = "\n")
    next(data)
    count = 0
    for i in enumerate(data_csv):
        count += 1
        for j in i[1]:
            cursor.execute('insert into carbon_nanotubes.carbon values ({data})'.format(data=(j)))
            mydb.commit()
    print("all data inserted")
    print(count)
```
